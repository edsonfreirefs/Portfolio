# 2 How to open an excel file using xlrd and Pandas
<hr>
The xlrd package is utilized for reading the sheet names.
<p>
The Pandas package is utilized for reading the tables into Data frames

```python
# package needed
import xlrd

# open excel file
xls = xlrd.open_workbook('Supplementary material/2 Limnology/Survey_campaigns.xlsx', on_demand=True)

# read sheet names and print
sheets = xls.sheet_names()
print(sheets)
```

```python
# package needed
import pandas as pd

# open the sheets in different pandas data frames; then, ts saves each dataframe in a dictionary
campaigns = {}
for x in sheets:
    
    data = pd.read_excel('Supplementary material/2 Limnology/Survey_campaigns.xlsx', sheet_name=x)
    campaigns[x] = data
```

# 3 The table data
<hr>
The Data Frame are limnological measurements take into four reservoirs, Ibitinga, Itaipu, Tres Marias, and Tucurui. In each Data Frame, the measurements taken are Chl-a, Pheophytin, Total Suspended Matter (TSM), Dissolved Organic Carbon (DOC), Secchi Depth (Zsd), and Temperature.

```python
# print the first 3 lines of each data frame
from IPython.core import display as ICD
for x in campaigns.keys():
    print(x)
    ICD.display(campaigns[x])
    print('\n\n')
```

# 4 How to concatenate all Data Frames into a single one

```python
# generate a new colums in each data frame, called reservoir, and then assign the reservoir name to each row
for x in campaigns.keys():
    
    campaigns[x]['Reservoir'] = x
```

```python
# concatenate all tables into a single table
first_campaign = list(campaigns.keys())[0]
table = campaigns[first_campaign]
for x in list(campaigns.keys())[1:]:
    
    table = table.append(campaigns[x])
```

# 5 Useful tools for extracting preliminary insights from data
<hr>

The Seaborn package is very useful for extracting some insights from data.
<p>
First, we can see how each parameter varies among the Reservoirs.

```python
# package needed
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

i = 1
plt.figure(figsize=(16,16))
for param in ['Chl-a (µg/L)', 'Pheophytin (µg/L)', 'TSM (mg/l)', 'DOC (mg/L)', 'Zsd (m)', 'Temperature (ºC)']:
    ax = plt.subplot(3,3, i)
    sns.boxplot(x='Reservoir', y=param, data=table, ax=ax)
    ax.grid(axis='y')
    ax.set_axisbelow(True)
    i = i + 1
    
plt.tight_layout()
```

We can also analysis how the parameters are correlated by reservoir.

```python
sns.pairplot(table.loc[:, ['Chl-a (µg/L)', 'Pheophytin (µg/L)', 'TSM (mg/l)', 'DOC (mg/L)', 'Zsd (m)', 'Reservoir']],
             hue='Reservoir')
```

And then calculate the Person correlation among variables for each reservoir.

```python
for x in campaigns.keys():
    print(x)
    subset = table[table['Reservoir'] == x].loc[:, ['Chl-a (µg/L)', 'Pheophytin (µg/L)', 'TSM (mg/l)', 'DOC (mg/L)', 'Zsd (m)']]
    ICD.display(subset.corr())
    print('\n \n')
```
