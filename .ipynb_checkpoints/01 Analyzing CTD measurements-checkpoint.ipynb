# 3 Create a function to read the CTD files
<hr>
<div style="text-align: justify">
The first necessary method is the readlines(), which read all lines in the document. Then, we can use a for loop to find the rows with the necessary data using the keywords (e.g., [Data]). Using the known rows, we can extract the id, date, and time data from the lines read, and also use the Pandas package for reading the table data.
</div>

```python
# Library utilized
import pandas as pd
import datetime as dt

# Function
def read_ctd_files(ctd_file):
    
    # Open the CDT file
    data = open(ctd_file, encoding='utf-8')
    lines = data.readlines()

    # Create a list with the rows of id, date, time, and the first row of the table data
    rows = []

    # for loop for each keyword
    for x in ['Block No', 'StartDate', 'StartTime', '[Data]']:

        # for loop for finding the row number of each keyword
        row_n = 1
        for y in lines:
            if x in y:
                rows.append(row_n)
                break
            row_n = row_n + 1

    # Reads the id
    id_data = int(lines[rows[0]-1].replace('Block No=', '').replace('\n', ''))

    # Reads the date
    date_data = lines[rows[1]-1].replace('StartDate=', '').replace('\n', '')

    # Reads the time
    time_data = lines[rows[2]-1].replace('StartTime=', '').replace('\n', '')

    # Combines Date and Time to date time
    date_time_data = date_data + ' ' + time_data
    date_time_data = dt.datetime.strptime(date_time_data, "%Y/%m/%d %H:%M:%S")

    # Reads the table data
    table_data = pd.read_csv(ctd_file, skiprows=rows[3])

    # Selects the desired parameters
    table_data = table_data.loc[:,['Depth', 'Salinity', 'Temp','Chl', 'Turbidity']]
    table_data['Depth'] = table_data['Depth']*-1
    
    # Defines output
    output = [id_data, date_time_data, table_data]
    
    return output
```

# 4 Read and plot all CTD files
<hr>

```python
# Library used
import os

# The path with the CTD files
ctd_path = 'Supplementary material/1 CTD'

# A list with the CTD files
ctd_files = os.listdir(ctd_path)

# Reads the CTD files into a dictionary
output = {}
for x in ctd_files:
    
    ctd = read_ctd_files(ctd_path+'/'+x)
    output[ctd[0]] = ctd[1:]

print('The CTD ids')
print(output.keys())
```

```python
# Library used
import matplotlib.pyplot as plt
from IPython.core import display as ICD

# Plot the profile of all parameters
n_subplots = len(output.keys())
i = 1
y_text = 1
fig = plt.figure(figsize=(16,40))

# For loop for each CTD ID
for x in output.keys():
    
    # Salinity plot
    ax = plt.subplot(n_subplots, 4, i)
    output[x][1].plot(y='Depth',
                      x='Salinity',
                      ax=ax,
                      xlim=(0,30),
                      ylim=(-6,0),
                      legend=False,
                      color='k')
    ax.set_ylabel('Depth (m)',
                  fontsize=12)
    ax.set_xlabel('Salinity (PSU)',
                  fontsize=12)
    plt.grid()
    
    # Temperature plot
    ax = plt.subplot(n_subplots, 4, i+1)
    output[x][1].plot(y='Depth',
                      x='Temp',
                      ax=ax,
                      xlim=(20,26),
                      ylim=(-6,0),
                      legend=False,
                      color='k')
    ax.set_ylabel('Depth (m)',
                  fontsize=12)
    ax.set_xlabel('Temperature (C¬∞)',
                  fontsize=12)
    plt.grid()
    
    # Turbidity plot
    ax = plt.subplot(n_subplots, 4, i+2)
    output[x][1].plot(y='Depth',
                      x='Turbidity',
                      ax=ax,
                      xlim=(0,10),
                      ylim=(-6,0),
                      legend=False,
                      color='k')
    ax.set_ylabel('Depth (m)',
                  fontsize=12)
    ax.set_xlabel('Turbidity (NTU)',
                  fontsize=12)
    plt.grid()
    
    # Chl-a concentration plot
    ax = plt.subplot(n_subplots, 4, i+3)
    output[x][1].plot(y='Depth',
                      x='Chl',
                      ax=ax,
                      xlim=(0,10),
                      ylim=(-6,0),
                      legend=False,
                      color='k')
    ax.set_ylabel('Depth (m)',
                  fontsize=12)
    ax.set_xlabel('Chl-a (mg/m¬≥)',
                  fontsize=12)
    plt.grid()
    
    # Change subplot row
    i = i + 4
    
plt.tight_layout(w_pad=1, h_pad=5)
  
# Add sampling title
for x in output.keys():
    fig.text(0.5,
             y_text,
             'Sampling Point '+ str(x),
            fontsize=16,
            verticalalignment='bottom',
            horizontalalignment='center')
    
    y_text = y_text - (1/n_subplots) - 0.001
```
